Importing Libraries

```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
from requests_html import HTMLSession
import textwrap
```

Fetching Links

```python
url = 'https://readnovelfull.com/ajax/chapter-archive?novelId=166'
response = requests.get(url)
soup = BeautifulSoup(response.content,'html5lib')
print(soup.prettify())
```

```python
links = []
url = 'https://readnovelfull.com{}'
table = soup.find('div',attrs = {'class':'panel panel-default'})
for link in table.find_all("a"):
    if link.has_attr('href'):
      url_links = url.format(link['href'])
      links.append(url_links)

for i in links:
  print(i)
```

Fetching Contents of Chapters and saving into text file

```python
session = HTMLSession()
```

```python
no = 1
for i in links:
  r = session.get(i)
  txt = r.html.find('#chr-content',first = True).text
  new_txt = textwrap.indent(txt, ' ', lambda line: True)
  postString = new_txt.replace("&apos","'")
  file_name = 'Chapter_{}.txt'
  file_name = file_name.format(no)
  file = open(file_name, 'w',encoding="utf-8")
  file.write(postString)
  file.close()

  with open(file_name,'r+',encoding="utf-8") as f:
    new_file = f.readlines()
    f.seek(0)
    for line in new_file:
      if "adsbygoogle" not in line:
        f.write(line)
    
    f.truncate()
  no = no +1
```
